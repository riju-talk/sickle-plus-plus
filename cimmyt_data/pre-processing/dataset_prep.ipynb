```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

nue = pd.read_csv(r'C:\Users\abhay\OneDrive\IIITD\Sem8\SSSE\Project\Data prepro\NUE_survey_dataset.csv')
odisha_supp = pd.read_excel(r'C:\Users\abhay\OneDrive\IIITD\Sem8\SSSE\Project\Data prepro\Odisha_Kharif2018.xlsx', sheet_name='Sheet1')

# Filter out the rows where region is "Odisha" in nue 
nue_odisha = nue[nue['Region'] == 'Odisha']

nue_odisha = nue_odisha[nue_odisha['SEASON'] == 'Rabi']

# Create new column in odisha_supp which rounds the value in Latitude and Longitude to 2 decimal places
odisha_supp['Latitude_round'] = odisha_supp['Latitude'].round(2)
odisha_supp['Longitude_round'] = odisha_supp['Longitude'].round(2)
```

```python
# Add a unique ID column to odisha_supp
odisha_supp = odisha_supp.reset_index().rename(columns={'index': 'id'})

# Create a mapping of rounded lat-long to list of ids
lat_long_to_ids = odisha_supp.groupby(['Latitude_round', 'Longitude_round'])['id'].apply(list).to_dict()

# Add a new column in nue_odisha with matching ids
nue_odisha['matched_ids'] = nue_odisha.apply(
    lambda row: lat_long_to_ids.get((row['LAT'], row['LONG']), []),
    axis=1
)
```

```python
nue_odisha['matched_ids']
```

```python
# Convert acres to hectares in odisha_supp
odisha_supp['C-q306_cropLarestAreaHectare'] = odisha_supp['C-q306_cropLarestAreaAcre'] / 2.47105

# Function to find the best match based on area
def find_best_match(row):
    matched_ids = row['matched_ids']
    if not matched_ids:  # If no matches, return None
        return None, None

    # Extract matched rows from odisha_supp
    matched_rows = odisha_supp.loc[odisha_supp['id'].isin(matched_ids), ['id', 'C-q306_cropLarestAreaHectare']]
    
    # Find the row with the minimum absolute area difference
    best_match = matched_rows.iloc[(matched_rows['C-q306_cropLarestAreaHectare'] - row['CRLPARHA']).abs().argmin()]
    
    return best_match['id'], abs(best_match['C-q306_cropLarestAreaHectare'] - row['CRLPARHA'])

# Apply the function to find best matches
nue_odisha[['chosen_mapping_id', 'mapping_area_difference']] = nue_odisha.apply(
    lambda row: pd.Series(find_best_match(row)), axis=1
)

# Display the updated dataframe
print(nue_odisha[['LAT', 'LONG', 'CRLPARHA', 'chosen_mapping_id', 'mapping_area_difference']])
```
