# We need earliest to 2016
for all regions but for baseline we do it on AP

```python
# SDATE is a date string (e.g. '2018-06-25' or dd/mm/yyyy) — parse it and get year
nue['SDATE'] = pd.to_datetime(nue['SDATE'], format='mixed', errors='coerce')
nue['SDATE_year'] = nue['SDATE'].dt.year

# Filter out the rows where year is less than 2016
nue = nue[nue['SDATE_year'] >= 2016]

# Filter out the rows where SDATE_rel is less than 0
nue = nue[nue['SDATE_rel'] >= 0]

# Date range of SDATE (after filtering)
print("SDATE range:", nue['SDATE'].min(), "to", nue['SDATE'].max())
```

```python
# TDATE is a date string — parse it and get year
nue['TDATE'] = pd.to_datetime(nue['TDATE'], format='mixed', errors='coerce')
nue['TDATE_year'] = nue['TDATE'].dt.year

# Filter out the rows where year is less than 2016
nue = nue[nue['TDATE_year'] >= 2016]

# Filter out the rows where TDATE_rel is less than 0
nue = nue[nue['TDATE_rel'] >= 0]

# Date range of TDATE (after filtering)
print("TDATE range:", nue['TDATE'].min(), "to", nue['TDATE'].max())
```

```python
# HDATE is a date string — parse it and get year
nue['HDATE'] = pd.to_datetime(nue['HDATE'], format='mixed', errors='coerce')
nue['HDATE_year'] = nue['HDATE'].dt.year

# Filter out the rows where year is less than 2016
nue = nue[nue['HDATE_year'] >= 2016]

# Date range of HDATE (after filtering)
print("HDATE range:", nue['HDATE'].min(), "to", nue['HDATE'].max())
```
